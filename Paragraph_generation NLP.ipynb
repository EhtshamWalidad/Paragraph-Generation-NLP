### Article Generation using N-grams

You have to write the urdu article using ngrams (1 grams to 5 grams). So, in short your output must be of 5 paragraphs, the first one is generated using unigram, second one is generated using bigram and so on.

Input: Your input is the seed sentence. E.g. first 3 to 4 words of the paragraph.

Output: Your output is the consist of 5 paragraphs for each n gram, each of 200 words.

You have to make N-gram model using the provided dataset. Dataset can be downloaded from  https://www.kaggle.com/datasets/saurabhshahane/urdu-news-dataset

You have to use all News Text column.

```python
#Your code here
import pandas as pd
data=pd.read_excel('urdu-news-dataset-1M.xlsx',nrows=15) #reading excel
urdu=pd.DataFrame(data)
urdu
```

# NGrams Code

```python
def create_ngrams(tokens,till):  #function to create ngrams list
    ngrams_listing=[]
    for sentence in tokens:
        for word in range(0, len(sentence)):
            gram =' '.join(sentence[word:word + int(till)])  #slicing
            ngrams_listing.append(gram)
    return ngrams_listing
```

# Tokenize and Probability

```python
from nltk.tokenize import word_tokenize
news=urdu['News Text']
tokens=[]
for i in range(0, len(news)):
    tokens.append(word_tokenize(str(news[i])))
count=1
dict={}
for i in news:
    print("Lines Read: ",count)
    w=str(i).split()
    count=count+1
    for j in w:
        j=j.lower()
        if j not in dict:
            dict[j]=1
        else:
            dict[j]+=1
total=sum(dict.values())
new_dict={}
l=0
for i,j in sorted(dict.items()):
    prob=j/total
    new_dict[list(dict.keys())[l]]=prob
    l=l+1
print(new_dict.items())
```

```python
import random
create=input("enter word to create paragraph: ")
word_limit=input("Enter Word Limit of one Paragraph: ")
print('\n')
create=create.split(" ")
create=create[len(create)-1]
print(create)             #check only the last words of the entered sentence
para=[]
count=1
count_down=int(word_limit)*5  #count down
ngrams=create_ngrams(tokens,1)
for n in range(0,len(ngrams)-(1)):  #for unigram cause it have only one in list (1)
        ran=random.choice(ngrams)
        para.append(ran)
        count=count+1
        count_down=count_down-1
        print("Time Left: ",count_down)
        create=ran
        if int(count)>=int(word_limit):  #condition of the word limit
                break;
for i in range(2,6):   #for 2,3,4,5 ngrams
    count=1
    para.append("   ")
    ngrams=create_ngrams(tokens,i)
    for n in range(0,len(ngrams)-(i-1)):
        m=[]                             #storing all the possible matching of words
        name=[]                          #storing names of matching
        prob=[]                          # storing probability of matching
        for i in range(0,len(ngrams)-1):
            if ngrams[i].split(" ")[0] == create:
                m.append(ngrams[i])
        for t in m:
            t=str(t).split(" ")
            p=0
            n_dict={}
            for i,j in new_dict.items():
                if i==t[len(t)-1]:
                        name.append(i)
        if name!=[]:
            name = list( dict.fromkeys(name))
            for na in name:
                p=0
                for i,j in new_dict.items():
                    if i==na:
                        p=p+j
                prob.append(p)
        if name!=[] and prob!=[]:
                ran=random.choice(prob)
                ind=prob.index(ran)
                para.append(name[ind])
                count=count+1
                count_down=count_down-1
                print("Time Left: ",count_down)
                create=(name[ind])
                if int(count)>=int(word_limit):
                    break;


new=' '.join(para)
new=new.split("   ")
for i in new:
    print(i)
    print('\n')
```
